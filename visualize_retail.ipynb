```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use("seaborn-darkgrid")
```

```python
data = "../datasets/BigBasketProducts.csv"
retail = pd.read_csv(data, index_col="index")
retail.dropna()
```

```python
# what brand has the higest store goods .
graph = retail["brand"].value_counts().head(2000)
graph.sort_values(na_position="first").plot(kind= "kde", figsize=(16,8)) #Density graph representation.
```

```python
graph.sort_values().plot(kind= "area", figsize=(18,10), xlabel="BRANDS", ylabel="ITEMS")
```

```python
retail["average_price"] = (retail.sale_price + retail.market_price)/2
retail["center_price"] = retail.average_price.map(lambda p: p - retail.average_price.mean()) 
retail
```

```python
retail.brand.describe()
```

```python
fruits= retail.loc[(retail.sub_category == "Dry Fruits") & (retail.average_price <= 500)] #checking Dry fruits less than or equals $100 .
```

```python
fruits.to_csv("fruit.csv")
fruits.groupby("product")["market_price"].sum().head(500).plot(kind="bar", color="orange", ylabel="Frequency(item-price)", title="Product Market-Price")
```

```python
medical = retail.loc[(retail.sub_category == 'Health & Medicine') & (retail.type == "Supplements & Proteins")]
# df.category.unique()
medical = medical.loc[(medical.brand == "ORGANIC INDIA") & (medical.rating <= 6)]
medical
```

```python
plt.figure(figsize=(16,8))
plt.plot(medical["average_price"],label= "Average Price", color="blue")
plt.plot(medical["sale_price"], label= "Center Price", color="red")


plt.legend(title = "Group")
plt.title("Price By Group", fontsize= 16)
plt.ylabel("Price", fontsize = 14)
plt.xlabel("Index", fontsize = 14)
```
